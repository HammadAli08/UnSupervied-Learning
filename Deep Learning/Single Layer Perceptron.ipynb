```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/home/hammadali08/Vidoes/CSV file/Job Placement Dataset.csv")
data.dropna(inplace=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
lbl.fit(data['status'])
data['status']=lbl.transform(data['status'])
```

```python
sns.scatterplot(x="ssc_percentage", y="emp_test_percentage", hue="status", data=data)
plt.show()
```

```python
x=data[['ssc_percentage','emp_test_percentage']]
y=data['status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr=Perceptron(l1_ratio=1)
pr.fit(x_train,y_train)
```

```python
pr.score(x_test,y_test),pr.score(x_train,y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
```
