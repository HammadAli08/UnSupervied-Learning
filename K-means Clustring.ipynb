```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL.ImageColor import colormap

data1=pd.read_csv("/home/hammadali08/Vidoes/CSV file/iris.csv")
```

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/iris.csv")
data.drop(columns=['species'],inplace=True)
data
```

```python
sns.pairplot(data)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,20):
    km=KMeans(n_clusters=i)
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot([i for i in range(1,20)],wcss,marker="o")
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.grid(axis="x")
plt.xticks([i for i in range(1,20)])
plt.show()
```

```python
km=KMeans(n_clusters=3)
data['Predict']=km.fit_predict(data)
data
```

```python
sns.pairplot(data,hue="Predict")
plt.savefig("/home/hammadali08/Vidoes/CSV file/iris.png")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data,data['Predict'])
```

```python
ss=[]
for i in range(2,20):
     km=KMeans(n_clusters=i)
     km.fit_predict(data)
     ss.append(silhouette_score(data,km.labels_))
```

```python
plt.bar([i for i in range(2,20)],ss)
plt.grid(axis='x')
plt.xticks([i for i in range(2,20)])
plt.show()
```

```python
plt.plot([i for i in range(2,20)],ss,marker="o")
plt.title("Sihlouette Score")
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Scores")
plt.xticks([i for i in range(2,20)])
plt.grid(axis="y")
plt.show()
```
