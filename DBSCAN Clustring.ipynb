```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.datasets import  make_moons
```

```python
x,y=make_moons(n_samples=300,noise=0.05)
```

```python
df=pd.DataFrame({'x':x[:,0],'y':x[:,1],'Output':y})
df
```

```python
sns.scatterplot(x='x',y='y',hue='Output',data=df)
```

```python
db=DBSCAN(eps=0.2,min_samples=2,metric = "euclidean")
```

```python
df['Predict']=db.fit_predict(x)
df
```

```python
scatter_plot=sns.scatterplot(x='x',y='y',hue='Predict',data=df)
```
