```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/home/hammadali08/Vidoes/CSV file/association example.csv')
data=data.head(10000)
data
```

```python
market=[]
for i in range(0,data.shape[0]):
    cus=[]
    for j in data.columns:
        if type(data[j][i])==str:
            cus.append(data[j][i])
        market.append(cus)
market
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
l
```

```python
import collections
p=collections.Counter(l)
df=pd.DataFrame({'List Names':p.keys(),'Count':p.values()})
df.sort_values('Count',ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
tr=TransactionEncoder()
```

```python
tr.fit(market)
```

```python
df1=pd.DataFrame(tr.transform(market),columns=tr.columns_)
df1
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fp=(fpgrowth(df1,min_support=0.07,use_colnames=True,max_len=3)).sort_values('support',ascending=False)
fp
```
