# Agglomertive CLustring

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/iris.csv")
data.drop(columns=['species'],inplace=True)
data
```

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sch.dendrogram(sch.linkage(data,method='average',metric='euclidean'))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage='complete')
```

```python
data['Predict']=(ac.fit_predict(data))
data
```

```python
sns.scatterplot(data,hue='Predict',x='sepal_length',y='sepal_width')
```

```python
sns.pairplot(data,hue='Predict',palette='Set2')
plt.show()
```
